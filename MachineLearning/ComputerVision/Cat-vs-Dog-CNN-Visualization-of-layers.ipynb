```python
def split_data(SOURCE, TRAINING, TESTING, SPLIT_SIZE):
    # YOUR CODE STARTS HERE
    dir_contents = os.listdir(SOURCE)
    #     full_paths = [os.path.join(SOURCE, file_name) for file_name in dir_contents]

    dir_contents = random.sample(dir_contents, len(dir_contents))
    dir_contents = [
        file_name
        for file_name in dir_contents
        if os.path.getsize(os.path.join(SOURCE, file_name))
    ]

    SPLIT_INDEX = round(len(dir_contents) * SPLIT_SIZE)

    training_files = dir_contents[:SPLIT_INDEX]
    testing_files = dir_contents[SPLIT_INDEX:]

    for training_file in training_files:
        copyfile(
            os.path.join(SOURCE, training_file), os.path.join(TRAINING, training_file)
        )

    for testing_file in testing_files:
        copyfile(
            os.path.join(SOURCE, testing_file), os.path.join(TESTING, testing_file)
        )
```

```python
CAT_SOURCE_DIR = "full_dataset/PetImages/Cat/"
TRAINING_CATS_DIR = "full_dataset/cats-v-dogs/training/cats/"
TESTING_CATS_DIR = "full_dataset/cats-v-dogs/testing/cats/"
DOG_SOURCE_DIR = "full_dataset/PetImages/Dog/"
TRAINING_DOGS_DIR = "full_dataset/cats-v-dogs/training/dogs/"
TESTING_DOGS_DIR = "full_dataset/cats-v-dogs/testing/dogs/"
```

```python
TRAIN_DATASET_PATH = "full_dataset/cats-v-dogs/training/"
VALIDATION_DATASET_PATH = "full_dataset/cats-v-dogs/testing/"
```

```python
TARGET_IMG_SIZE = 150
BATCH_SIZE = 32
```

```python
train_data_gen = ImageDataGenerator(rescale=1.0 / 255)

train_generator = train_data_gen.flow_from_directory(
    TRAIN_DATASET_PATH,
    target_size=(TARGET_IMG_SIZE, TARGET_IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
)
```

```python
validation_data_gen = ImageDataGenerator(rescale=1.0 / 255)

validation_generator = validation_data_gen.flow_from_directory(
    VALIDATION_DATASET_PATH,
    target_size=(TARGET_IMG_SIZE, TARGET_IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
)
```

```python
def create_conv_model() -> keras.Model:

    model = keras.Sequential()

    model.add(
        keras.layers.Conv2D(
            filters=16,
            kernel_size=(3, 3),
            activation="relu",
            input_shape=(TARGET_IMG_SIZE, TARGET_IMG_SIZE, 3),
        )
    )
    model.add(keras.layers.MaxPool2D(pool_size=(2, 2)))

    model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation="relu"))
    model.add(keras.layers.MaxPool2D(pool_size=(2, 2)))

    model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation="relu"))
    model.add(keras.layers.MaxPool2D(pool_size=(2, 2)))

    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(units=512, activation="relu"))
    model.add(keras.layers.Dense(units=128, activation="relu"))
    model.add(keras.layers.Dense(units=32, activation="relu"))
    model.add(keras.layers.Dense(units=1, activation="sigmoid"))

    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

    return model
```

```python
conv_model = create_conv_model()
```

```python
history = conv_model.fit_generator(
    train_generator,
    #     steps_per_epoch=90,
    epochs=15,
    validation_data=validation_generator,
    #     validation_steps=9,
)
```

```python
plt.plot(history.epoch, history.history["loss"], color="r")
plt.plot(history.epoch, history.history["val_loss"], color="g")

plt.title("Loss Train vs Validation")
```

```python
plt.plot(history.epoch, history.history["accuracy"], color="r")
plt.plot(history.epoch, history.history["val_accuracy"], color="g")

plt.title("Accuracy Train vs Validation")
```

```python
conv_model.evaluate_generator(validation_generator)
```

## Visualizing intermediate representation

```python
cat_image_dir = os.path.join(TRAIN_DATASET_PATH, "cats")
dog_image_dir = os.path.join(TRAIN_DATASET_PATH, "dogs")

cat_image_paths = [
    os.path.join(cat_image_dir, file_name) for file_name in os.listdir(cat_image_dir)
]
dog_image_paths = [
    os.path.join(dog_image_dir, file_name) for file_name in os.listdir(dog_image_dir)
]
```

```python
def get_random_image():

    random_img = random.choice(cat_image_paths + dog_image_paths)
    img = keras.preprocessing.image.load_img(
        random_img, target_size=(TARGET_IMG_SIZE, TARGET_IMG_SIZE)
    )

    print(random_img)
    plt.imshow(img)

    img = np.expand_dims(img, axis=0)
    img = np.vstack(img)

    img = img.reshape(1, img.shape[0], img.shape[1], img.shape[2])
    return img
```

```python
sample_image = get_random_image()
```

```python
successive_layers = [layer.output for layer in conv_model.layers]
```

```python
visualization_model = keras.models.Model(
    inputs=conv_model.input, outputs=successive_layers
)
```

```python
def visualize_intermediate_layers():

    sample_image = get_random_image()
    prediction_class = conv_model.predict(sample_image)

    if prediction_class[0] > 0:
        print("It's a Dog")
    else:
        print("It's a cat")

    feature_maps = visualization_model.predict(sample_image)

    layer_names = [layer.name for layer in conv_model.layers]

    for layer_name, feature_map in zip(layer_names, feature_maps):

        n_features = feature_map.shape[-1]
        feature_map_size = feature_map.shape[1]

        if len(feature_map.shape) == 4:

            display_grid = np.zeros((feature_map_size, feature_map_size * n_features))

            for i in range(n_features):
                x = feature_map[0, :, :, i]
                x -= x.mean()
                x /= x.std()
                x *= 64
                x += 128
                x = np.clip(x, 0, 255)
                display_grid[:, i * feature_map_size : (i + 1) * feature_map_size] = x

            scale = 40.0 / n_features
            plt.figure(figsize=(scale * n_features, scale))

            plt.title(layer_name)
            plt.grid(False)
            plt.imshow(display_grid, aspect="auto", cmap="viridis")
```

```python
visualize_intermediate_layers()
```
