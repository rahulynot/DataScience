```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
import matplotlib.pyplot as plt

%matplotlib inline
%reload_ext nb_black
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

## Classification

```python
tree_clf = DecisionTreeClassifier(max_depth=3)
tree_clf.fit(X, y)
```

```python
plt.figure(figsize=(20, 15))

plot_tree(
    tree_clf,
    filled=True,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
)

plt.show()
```

```python
print(tree_clf.predict([X[50]]))
print(y[50])
```

```python
tree_clf.predict_proba([X[50]])
```

## Regression

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston

X, y = load_boston(return_X_y=True)
feature_names = load_boston().feature_names
```

```python
tree_reg = DecisionTreeRegressor(max_depth=3, min_samples_leaf=3)
tree_reg.fit(X, y)
```

```python
plt.figure(figsize=(20, 15))

plot_tree(tree_reg, filled=True, feature_names=feature_names)

plt.show()
```

```python
print(tree_reg.predict([X[50]]))
print(y[50])
```
